```python
import numpy as np
import pylab

x = np.arange(-2*np.pi, 2*np.pi, 0.001)
y = np.cos(x)

pylab.figure(figsize=(15, 6))

pylab.plot(x, y, 'b-o', label="cos")
```

```python
np.random.seed(1)

bruit = np.random.normal(loc=0, scale=0.8, size=y.shape[0])

pylab.plot(x, bruit)
```

```python
signalB = y + bruit


pylab.figure(figsize=(15, 6))


pylab.plot(x, signalB, 'r-', label="signal bruite")
#pylab.plot(x, y, 'b-', label="signal non bruite")

pylab.legend()
```

```python
taille = [10, 20, 50, 1000]#, 50, 1000]

pylab.figure(figsize=(15, 6))

pylab.plot(x,signalB, label="signal non filtre")
for w in taille:
  poids = np.ones(w)/w
  signalFiltre = np.convolve(signalB, poids, 'valid')
  pylab.plot(x[w-1:], signalFiltre, label="signal filtre %d"%w)


pylab.plot(x,y,'r', label="cosine")
pylab.legend()
```
